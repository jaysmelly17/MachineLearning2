## Section 1: What Is a List?

A **list** is an ordered collection of items. It's one of the most important data structures in Python — you'll use lists constantly.

Think of a list like a numbered row of boxes. Each box holds one value, and each box has a position number called an **index**. Indexes start at **0**, not 1.

```
Index:    0         1        2        3
       ┌─────┐  ┌──────┐  ┌───────┐ ┌──────┐
       │"red"│  │"blue"│  │"green"│ │"gold"│
       └─────┘  └──────┘  └───────┘ └──────┘
```

**Syntax:** Square brackets `[]` with items separated by commas. A list can hold any data type — strings, ints, floats, even other lists.

````python
# EXAMPLE: Creating lists and accessing items by index

# A list of strings
colors = ["red", "blue", "green", "gold"]
print(colors)              # prints the whole list: ['red', 'blue', 'green', 'gold']

# Access items by index — remember, indexes start at 0!
print(colors[0])           # prints: red (first item)
print(colors[2])           # prints: green (third item)
print(colors[-1])          # prints: gold (last item — negative indexes count from the end)

# Lists can hold different data types
mixed = ["Alice", 17, 3.8, True]   # string, int, float, bool — all in one list
print(mixed)

# Check how many items are in a list with len()
print(f"colors has {len(colors)} items")   # prints: colors has 4 items
````

## Section 2: Modifying Lists

Lists are **mutable** — you can change them after they're created. Here are the key operations:

| Operation | What It Does | Syntax |
|-----------|-------------|--------|
| Change an item | Overwrite by index | `myList[i] = newValue` |
| Add to the end | Append one item | `myList.append(item)` |
| Insert at position | Shifts others right | `myList.insert(i, item)` |
| Remove by value | Removes first match | `myList.remove(item)` |
| Remove by index | Removes at position | `del myList[i]` |
| Remove & return last | Pop from end | `myList.pop()` |

````python
# EXAMPLE: All the ways to modify a list

fruits = ["apple", "banana", "cherry"]
print(f"Start:   {fruits}")

# CHANGE an item by index
fruits[1] = "blueberry"              # replace index 1 ("banana") with "blueberry"
print(f"Change:  {fruits}")

# APPEND adds to the end of the list
fruits.append("dragonfruit")          # adds "dragonfruit" after the last item
print(f"Append:  {fruits}")

# INSERT adds at a specific position (other items shift right)
fruits.insert(0, "avocado")           # insert at index 0 — becomes the new first item
print(f"Insert:  {fruits}")

# REMOVE deletes the first occurrence of a value
fruits.remove("cherry")               # finds "cherry" and removes it
print(f"Remove:  {fruits}")

# DEL removes by index number
del fruits[0]                          # removes index 0 ("avocado")
print(f"Del [0]: {fruits}")

# POP removes and returns the last item
last = fruits.pop()                    # removes "dragonfruit" and stores it in 'last'
print(f"Popped:  {last}")
print(f"After:   {fruits}")
````

## Section 3: Looping Through Lists

The most common thing you'll do with a list is **loop through it** and do something with each item. Python makes this easy with a `for` loop.

There are two styles:
- **For-each loop:** `for item in myList:` — gives you the value directly (simpler, use this when you just need the values)
- **Index loop:** `for i in range(len(myList)):` — gives you the index number (use when you need to know the position or modify items)

````python
# EXAMPLE: Two ways to loop through a list

scores = [92, 87, 95, 78, 88]

# Style 1: For-each — just gives you each value
# 'score' is a new variable that takes on each value in the list, one at a time
print("=== For-each loop ===")
for score in scores:
    print(f"Score: {score}")

# Style 2: Index loop — gives you the index number
# range(len(scores)) generates 0, 1, 2, 3, 4
# Then you use the index to access the item
print("\n=== Index loop ===")
for i in range(len(scores)):          # i goes 0, 1, 2, 3, 4
    print(f"Index {i}: {scores[i]}")  # use i to look up the item

# Common pattern: calculate a total using a loop
total = 0                              # start a running total at 0
for score in scores:                   # loop through each score
    total += score                     # add it to the running total
average = total / len(scores)          # divide total by count
print(f"\nTotal: {total}, Average: {average:.1f}")
````

## Section 4: Slicing Lists

**Slicing** lets you grab a portion of a list — a sublist. The syntax is:

```python
myList[start:stop]       # from start index up to (but NOT including) stop index
myList[start:stop:step]  # with a step (skip every N items)
```

Key rules:
- `start` is **included**, `stop` is **excluded** (just like `range()`)
- If you omit `start`, it defaults to the beginning
- If you omit `stop`, it goes to the end

````python
# EXAMPLE: Slicing — grab parts of a list

nums = [10, 20, 30, 40, 50, 60, 70]
#        0   1   2   3   4   5   6    ← indexes

# Basic slicing: [start:stop] — start is included, stop is NOT
print(nums[1:4])        # [20, 30, 40] — indexes 1, 2, 3 (not 4)
print(nums[0:3])        # [10, 20, 30] — indexes 0, 1, 2

# Omit start → starts from beginning
print(nums[:3])         # [10, 20, 30] — first 3 items

# Omit stop → goes to the end
print(nums[4:])         # [50, 60, 70] — everything from index 4 onward

# Negative indexes — count from the end
print(nums[-3:])        # [50, 60, 70] — last 3 items

# Step — take every Nth item
print(nums[::2])        # [10, 30, 50, 70] — every 2nd item (indexes 0, 2, 4, 6)

# Reverse a list with step -1
print(nums[::-1])       # [70, 60, 50, 40, 30, 20, 10] — backwards
````

## Section 5: Useful List Functions

Python gives you a bunch of built-in functions and methods that work with lists. Here are the ones you'll use most:

| Function/Method | What It Does | Example |
|----------------|-------------|--------|
| `len(lst)` | Number of items | `len([1,2,3])` → `3` |
| `sum(lst)` | Total of all items | `sum([1,2,3])` → `6` |
| `min(lst)` | Smallest item | `min([5,2,8])` → `2` |
| `max(lst)` | Largest item | `max([5,2,8])` → `8` |
| `lst.sort()` | Sort in place (changes the list) | |
| `sorted(lst)` | Returns a new sorted list (original unchanged) | |
| `lst.count(x)` | How many times x appears | `[1,2,2,3].count(2)` → `2` |
| `x in lst` | True if x is in the list | `3 in [1,2,3]` → `True` |

````python
# EXAMPLE: Useful list functions and methods

grades = [88, 92, 75, 95, 88, 83, 91]

# Built-in functions
print(f"Count:   {len(grades)}")       # 7 items in the list
print(f"Sum:     {sum(grades)}")       # total of all grades
print(f"Min:     {min(grades)}")       # lowest grade: 75
print(f"Max:     {max(grades)}")       # highest grade: 95
print(f"Average: {sum(grades) / len(grades):.1f}")  # average: sum / count

# sorted() returns a NEW sorted list — original is unchanged
sorted_grades = sorted(grades)          # creates a new sorted list
print(f"\nSorted:   {sorted_grades}")
print(f"Original: {grades}")            # original is still in the same order

# .sort() modifies the list IN PLACE — no return value
grades.sort()                            # changes grades itself
print(f"After .sort(): {grades}")       # now grades IS sorted

# .count() — how many times a value appears
print(f"\n88 appears {grades.count(88)} times")

# 'in' operator — check if something is in the list
print(f"Is 95 in the list? {95 in grades}")    # True
print(f"Is 100 in the list? {100 in grades}")  # False
````

# PART B: DICTIONARIES

Lists are great for ordered collections, but they have a limitation: you access data by **position number**. If you have `student = ["Alice", 11, 3.8]`, you have to remember that index 0 is the name, index 1 is the grade, and index 2 is the GPA. That gets confusing fast.

**Dictionaries** solve this by letting you access data by **name** instead of by number.

## Section 6: What Is a Dictionary?

A **dictionary** stores data as **key-value pairs**. Instead of using index numbers (0, 1, 2...), you use meaningful names.

```python
# LIST — access by position number (what does index 2 mean? who knows)
student_list = ["Alice", 11, 3.8]
print(student_list[2])                # 3.8 — but you have to remember that's the GPA

# DICTIONARY — access by name (crystal clear)
student_dict = {"name": "Alice", "grade": 11, "gpa": 3.8}
print(student_dict["gpa"])            # 3.8 — obvious what you're getting
```

**Syntax:** Curly braces `{}` with `key: value` pairs separated by commas. Keys are usually strings.

````python
# EXAMPLE: Creating a dictionary and accessing values

# Create a dictionary with 4 key-value pairs
# Keys are on the left of the colon, values are on the right
student = {
    "name": "Alice",          # key is "name", value is "Alice"
    "grade": 11,              # key is "grade", value is 11
    "gpa": 3.8,               # key is "gpa", value is 3.8
    "email": "alice@mccc.edu" # key is "email", value is the email string
}

# Print the entire dictionary
print(student)

# Access a single value using its key inside square brackets
print(student["name"])        # prints: Alice
print(student["gpa"])         # prints: 3.8

# Use dictionary values inside an f-string
# Note: inside the f-string we use single quotes for the key
# because the f-string itself uses double quotes
print(f"{student['name']} is in grade {student['grade']} with a {student['gpa']} GPA")
````

## Section 7: Modifying Dictionaries

Like lists, dictionaries are **mutable**. You can change values, add new keys, and delete keys:

| Operation | Syntax | Example |
|-----------|--------|---------|
| Change a value | `d[key] = new_val` | `d["gpa"] = 3.9` |
| Add a new key | `d[new_key] = val` | `d["phone"] = "555"` |
| Delete a key | `del d[key]` | `del d["phone"]` |

**Difference from lists:** With a list, adding/removing items shifts all the index numbers around. With a dictionary, keys are stable — deleting `"phone"` doesn't affect `"name"` or `"gpa"`.

````python
# EXAMPLE: Modifying, adding, and deleting dictionary entries

student = {"name": "Alice", "grade": 11, "gpa": 3.8}
print(f"Original: {student}")

# CHANGE an existing value — Alice's GPA went up
# Same bracket syntax as accessing, just assign a new value
student["gpa"] = 3.95
print(f"After GPA change: {student}")

# ADD a brand new key that didn't exist before
# If the key doesn't exist, Python creates it automatically
student["phone"] = "555-9999"
print(f"After adding phone: {student}")

# DELETE a key-value pair entirely using del
del student["phone"]
print(f"After deleting phone: {student}")
````

## Section 8: Safe Access with .get()

Here's a gotcha with dictionaries: if you use `dict[key]` and the key doesn't exist, Python **crashes** with a `KeyError`. This can break your whole program.

The `.get()` method is the safe alternative. It lets you provide a **default value** to use when a key isn't found:

```python
dict.get(key, default_value)
```

- Key exists → returns the value (same as `dict[key]`)
- Key missing → returns the default value instead of crashing

````python
# EXAMPLE: .get() vs bracket notation

student = {"name": "Alice", "grade": 11, "gpa": 3.8}

# Bracket notation works fine when the key exists
print(student["name"])                      # prints: Alice

# But CRASHES when the key is missing — uncomment to see:
# print(student["phone"])                   # KeyError: 'phone'

# .get() is the safe alternative
# First argument: the key to look for
# Second argument: what to return if it's not found
phone = student.get("phone", "N/A")         # key missing → returns "N/A" (no crash)
print(f"Phone: {phone}")                     # prints: Phone: N/A

# .get() still works normally when the key DOES exist
name = student.get("name", "Unknown")        # key found → returns "Alice" as normal
print(f"Name: {name}")                        # prints: Name: Alice
````

## Section 9: Looping Through Dictionaries

Just like lists, you can loop through dictionaries. There are three methods:

| Method | What You Get | When To Use |
|--------|-------------|-------------|
| `.keys()` | Just the keys | When you only need the names |
| `.values()` | Just the values | When you only need the data |
| `.items()` | Both key AND value | Most common — when you need everything |

With `.items()`, you **unpack** the key and value into two variables: `for key, value in dict.items():`

````python
# EXAMPLE: Three ways to loop through a dictionary

student = {"name": "Alice", "grade": 11, "gpa": 3.8}

# Loop 1: .keys() — gives you just the key names
print("=== Keys Only ===")
for key in student.keys():           # key is "name", then "grade", then "gpa"
    print(key)

# Loop 2: .values() — gives you just the values
print("\n=== Values Only ===")
for value in student.values():       # value is "Alice", then 11, then 3.8
    print(value)

# Loop 3: .items() — the most useful! gives you both as a pair
# Two variables (key, value) are separated by a comma
print("\n=== Both (Key: Value) ===")
for key, value in student.items():
    print(f"{key}: {value}")
````

## Section 10: Lists of Dictionaries (Nested Data)

This is where lists and dictionaries combine into something really powerful. Think of a spreadsheet:

| name | grade | gpa |
|------|-------|-----|
| Alice | 11 | 3.8 |
| Bob | 12 | 3.5 |
| Charlie | 11 | 3.9 |

The whole table is a **list** (of rows). Each row is a **dictionary** (with named columns).

To access a specific value, you chain the list index and the dictionary key: `students[0]["name"]` → first go to row 0, then grab the "name" field.

````python
# EXAMPLE: A list of dictionaries — like a database table

# The list holds all the "rows"
# Each dictionary is one "row" with named fields
students = [
    {"name": "Alice", "grade": 11, "gpa": 3.8},
    {"name": "Bob", "grade": 12, "gpa": 3.5},
    {"name": "Charlie", "grade": 11, "gpa": 3.9}
]

# Access one row (it's a list, so use an index)
print(students[0])                # prints the entire first dictionary

# Access one specific field
# First [0] gets the dict at index 0, then ["name"] gets the name from that dict
print(students[0]["name"])        # prints: Alice
print(students[2]["gpa"])         # prints: 3.9

# Loop through all rows
# Each time through the loop, 's' holds one entire dictionary
print("\n--- All Students ---")
for s in students:
    print(f"{s['name']} - Grade {s['grade']} - GPA: {s['gpa']}")

# Calculate the average GPA
total = 0                          # running total
for s in students:                 # loop through each student dict
    total += s["gpa"]              # add this student's GPA to total
average = total / len(students)    # divide by number of students
print(f"\nAverage GPA: {average:.2f}")
````

## Section 11: Building a Dictionary from a List (Counting Pattern)

One of the most common real-world tasks is **counting how many times each item appears** in a list. Dictionaries are perfect for this — the key is the item and the value is the count.

**The pattern:**
1. Start with an empty dictionary `{}`
2. Loop through the list
3. For each item: if it's already a key, add 1. If not, set it to 1.

````python
# EXAMPLE: Counting items in a list using a dictionary

# A list with repeated items
fruits = ["apple", "banana", "apple", "cherry", "banana", "apple", "cherry"]

# Start with an empty dictionary — we'll build it up
counts = {}

# Loop through every item in the list
for fruit in fruits:
    if fruit in counts:        # is this fruit already a key in our dict?
        counts[fruit] += 1     # yes — add 1 to the existing count
    else:
        counts[fruit] = 1      # no — create the key and start the count at 1

# Print the result
print(counts)                   # {'apple': 3, 'banana': 2, 'cherry': 2}

# Print it nicely
for fruit, count in counts.items():
    print(f"{fruit}: {count}")
````
