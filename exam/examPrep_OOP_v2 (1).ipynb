## Section 1: Classes, Objects, and Methods

A **class** is a blueprint. An **object** is one specific thing built from that blueprint. Think of the class as a cookie cutter and each object as an actual cookie.

Every class has a special method called `__init__`. Python calls it **automatically** the moment you create a new object — it's where the object's starting data gets set up.

**What is `self`?** Every method in a class takes `self` as its first parameter. It's a reference to *this specific object*. When you have two Dog objects, `self` is how each one knows to use *its own* name and breed — not some other dog's. You never pass `self` yourself; Python handles it automatically.

**Attributes** are variables that belong to an object, always written as `self.something`. **Methods** are functions that belong to an object — they can use `self` to access the object's own attributes.

```python
class ClassName:                          # define the blueprint
    def __init__(self, param1, param2):   # runs automatically when you create an object
        self.attribute1 = param1          # store data as attributes using self.
        self.attribute2 = param2

    def some_method(self):                # a method — a function that belongs to the object
        print(self.attribute1)            # uses self. to access its own data

obj = ClassName("value1", "value2")       # create an object — __init__ runs here
obj.some_method()                         # call a method using dot notation
```

````python
# EXAMPLE: A Dog class

class Dog:
    def __init__(self, name, breed, age):  # called automatically when you create a Dog
        self.name = name                   # self.name is the attribute — it sticks around
        self.breed = breed                 # name/breed/age are just the incoming parameters
        self.age = age

    def bark(self):                        # method — uses self to access this dog's data
        print(f"{self.name} says: Woof!")

    def describe(self):
        print(f"{self.name} is a {self.age}-year-old {self.breed}.")


# Create two Dog objects from the same blueprint — each has its OWN data
dog1 = Dog("Buddy", "Golden Retriever", 3)   # __init__ runs, self = dog1
dog2 = Dog("Max", "German Shepherd", 5)       # __init__ runs again, self = dog2

dog1.describe()   # Buddy is a 3-year-old Golden Retriever.
dog2.describe()   # Max is a 5-year-old German Shepherd.
dog1.bark()       # Buddy says: Woof!

# You can read and change attributes directly
dog1.age = 4
print(f"After birthday: {dog1.name} is now {dog1.age}")
````

### Practice Problem 1

Create a class called `Phone` with:
- **Attributes** (in `__init__`): `brand`, `model`, `battery` (a number from 0–100)
- **Method** `describe()`: prints something like `"iPhone 15 (Apple) — Battery: 85%"`
- **Method** `charge()`: sets battery to 100 and prints `"Phone fully charged!"`

Then:
1. Create two different Phone objects
2. Call `describe()` on both
3. Call `charge()` on the one with lower battery, then call `describe()` again to confirm it changed

````python
class Phone:
    def __init__(self, brand, model, battery):
        self.brand = brand          # store brand as attribute
        self.model = model          # store model as attribute
        self.battery = battery      # store battery level (0-100)
    
    def describe(self):
        print(f"{self.model} ({self.brand}) — Battery: {self.battery}%")
    
    def charge(self):
        self.battery = 100          # set battery to full
        print("Phone fully charged!")

# Create two phones and test them
phone1 = Phone("Apple", "iPhone 15", 85)
phone2 = Phone("Samsung", "Galaxy S24", 42)

phone1.describe()     # iPhone 15 (Apple) — Battery: 85%
phone2.describe()     # Galaxy S24 (Samsung) — Battery: 42%

phone2.charge()       # Phone fully charged!
phone2.describe()     # Galaxy S24 (Samsung) — Battery: 100%
````

## Section 2: Methods That Return Values

Methods are just functions inside a class. Like any function, they can either **print** something or **return** a value — and knowing which to use matters.

- Use **`print`** when the method's job is just to display something. The caller doesn't get anything back.
- Use **`return`** when the caller needs a value to use in their own code — store it, do math with it, put it in an `if` statement.

```python
def describe(self):          # prints — caller gets nothing back
    print(self.name)

def get_name(self):          # returns — caller can use the value
    return self.name

# If get_name() used print instead of return:
greeting = "Hello, " + obj.get_name()   # this only works because get_name() RETURNS
```

A method that prints can't be used in expressions like that. If you're not sure which to use, ask: *"does the caller need to do something with this value?"* If yes → `return`.

````python
# EXAMPLE: A Rectangle class — some methods print, some return

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):                          # RETURNS a value — caller can use it
        return self.width * self.height

    def perimeter(self):                     # RETURNS a value
        return 2 * (self.width + self.height)

    def describe(self):                      # PRINTS — just displays info, returns nothing
        print(f"Rectangle {self.width}x{self.height}")
        print(f"  Area: {self.area()}")      # can call its OWN methods with self
        print(f"  Perimeter: {self.perimeter()}")


r1 = Rectangle(5, 3)
r2 = Rectangle(10, 7)

r1.describe()
r2.describe()

# Because area() RETURNS, we can use it in a calculation
total = r1.area() + r2.area()
print(f"\nTotal area: {total}")
````

### Practice Problem 2

Create a class called `BankAccount` with:
- **Attributes** (in `__init__`): `owner` (string), `balance` — set this to `0` inside `__init__`, don't pass it in as a parameter
- **Method** `deposit(amount)`: adds the amount to the balance and prints the new balance
- **Method** `withdraw(amount)`: if there's enough money, subtracts from balance and prints new balance. If not, prints `"Insufficient funds!"`
- **Method** `get_balance()`: **returns** the current balance (don't print — return it)

Then:
1. Create an account
2. Deposit $200
3. Withdraw $75
4. Try to withdraw $500 (should fail)
5. Print the final balance using `get_balance()`

````python
class BankAccount:
    def __init__(self, owner):
        self.owner = owner          # store the owner's name
        self.balance = 0            # every new account starts at $0
    
    def deposit(self, amount):      # takes an amount to add
        self.balance += amount      # add to balance
        print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")
    
    def withdraw(self, amount):     # takes an amount to remove
        if amount > self.balance:   # check if there's enough money
            print(f"Insufficient funds! Balance: ${self.balance:.2f}")
        else:
            self.balance -= amount  # subtract from balance
            print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
    
    def get_balance(self):          # returns the balance (doesn't print)
        return self.balance

# Test it
acct = BankAccount("Alice")
acct.deposit(100)        # Deposited $100.00. New balance: $100.00
acct.withdraw(30)        # Withdrew $30.00. New balance: $70.00
acct.withdraw(200)       # Insufficient funds! Balance: $70.00
print(f"Final balance: ${acct.get_balance():.2f}")  # Final balance: $70.00
````
